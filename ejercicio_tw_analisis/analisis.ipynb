```python
# Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bikes = pd.read_excel('Bikes.xlsx')
bikes
```

```python
bikeshops = pd.read_excel('bikeshops.xlsx')
bikeshops
```

```python
orderlines = pd.read_excel('orderlines.xlsx')
orderlines
```

```python
# Suponiendo que 'orderline' y 'bikes' son tus DataFrames
merged_df = pd.merge(orderlines, bikes, left_on='product.id', right_on='bike.id')
```

```python
merged_df
```

```python
data = pd.merge(merged_df, bikeshops, left_on='order.line', right_on='bikeshop.id')
```

```python
data
```

```python
data = data.drop(columns=['product.id', 'order.line'])
data
```

```python
data.isnull().sum()
```

```python
data['category'] = data['description'].apply(lambda x: x.split()[0] if x.split()[0] in ['Road', 'Mountain'] else None)
```

```python
data
```

```python
data['Revenue'] = data['quantity'] * data['price']
data
```

```python
# Filtro de Supersix
# Filtra las filas donde la primera palabra en 'model' es 'Supersix'
filtered_data = data[data['model'].apply(lambda x: str(x).split()[0] == 'Supersix')]

# Crea una nueva columna 'State' en 'filtered_data' extrayendo la última palabra de 'location'
filtered_data['State'] = filtered_data['location'].apply(lambda x: str(x).split()[-1] if pd.notnull(x) else None)

# Muestra el DataFrame filtrado con la nueva columna
print(filtered_data)
```

```python
newData = filtered_data[filtered_data['State'].apply(lambda x: str(x).split()[0] == 'FL')]
newData
```

```python
# Calcula los cuantiles 25% y 75% de la columna 'Revenue'
q25 = newData['Revenue'].quantile(0.25)
q75 = newData['Revenue'].quantile(0.75)

# Función para clasificar cada valor en 'Revenue'
def classify_revenue(value):
    if value <= q25:
        return 'Low'
    elif value <= q75:
        return 'Medium'
    else:
        return 'High'

# Aplica la función para crear la nueva columna 'Revenue Quantiles'
newData['Revenue Quantiles'] = newData['Revenue'].apply(classify_revenue)
```
